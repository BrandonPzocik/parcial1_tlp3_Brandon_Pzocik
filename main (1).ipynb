- con el metodo .to_datatime paramos los valores a tipo fecha y se usa el combine_first para combinar a un solo formato de fecha 

## 3.3
 - para capitalizar el nombre y apellido usamos las funciones de str .title() para pasar a mayuscula las primera letra de cada palabra y .strip() para eliminar espacion adelante y atras de la palabraa 


```python
df.dtypes
```

```python
df["Precio_Unitario"].fillna(df["Precio_Unitario"].median(), inplace= True)
df.head(20)
```

```python
df.isnull().sum()
```

```python
df["Total_Venta"].fillna(df["Total_Venta"].median(), inplace= True)
df.head()
```

```python
df.isnull().sum()
```

```python
df["Producto"].fillna("Producto sin nombre", inplace= True)
df.tail()
```

```python
df["Metodo_Pago"].fillna("Metodo sin especificar", inplace=True)
df.head(30)
```

```python
df.isnull().sum()
```

```python
df["Region"].fillna("region no especificada", inplace=True)
df.isnull().sum()
```

```python
fecha1 = pd.to_datetime(df["Fecha"], format="%Y-%m-%d", errors="coerce")
fecha2 = pd.to_datetime(df["Fecha"], format="%d-%m-%Y", errors="coerce")
df["Fecha"] = fecha1.combine_first(fecha2)
df.dtypes
```

```python
df["Nombre_Cliente"] = df["Nombre_Cliente"].astype(str).str.title().str.strip()
df["Apellido_Cliente"] = df["Apellido_Cliente"].astype(str).str.title().str.strip()
df.head(10)
```
